```python
sites = res_dd['siteID']
res_dir = '../../DATA/WUS/WUS_output/bf/files/_1'
k_list = ['s_wilt', 's_star', 'beta_ww', 'pi_F', 'pi_R', 'pi_T', 'pi_S', 'Pg50', 'Px50', 'RAI', 'k_xl_max']
print('var\t\tmeanCV\tmedCV\tminCV\tmaxCV')
for k in k_list:
    cv = []
    for ix, site in enumerate(sites):
        outfile_res = os.path.join(res_dir, '%s.pickle' % site)
        with open(outfile_res, 'rb') as f:
            loc_res = load(f)
        #cv.append(np.abs(np.std(loc_res['%s_estimates' % k]) / loc_res[k]))
        cv.append(np.abs(np.std(loc_res['%s_estimates' % k]) / np.mean(loc_res['%s_estimates' % k])))
    if k != 'k_xl_max':
        print('%s\t\t%-5.2f\t%-5.2f\t%-5.2f\t%-5.2f' % (k, np.mean(cv), np.median(cv), np.min(cv), np.max(cv)))
    else:
        print('%s\t%-5.2f\t%-5.2f\t%-5.2f\t%-5.2f' % (k, np.mean(cv), np.median(cv), np.min(cv), np.max(cv)))
```
